### How to turn a time series to a matrix 

```python
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
         # find the end of this pattern
         end_ix = i + n_steps_in
         out_end_ix = end_ix + n_steps_out
         # check if we are beyond the sequence
         if out_end_ix > len(sequence):
             break
         # gather input and output parts of the pattern
         seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
         X.append(seq_x)
         y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
# choose a number of time steps
n_steps_in, n_steps_out = 10,5

# split into samples
X, y = split_sequence(series, n_steps_in, n_steps_out)
count = 0
for i in range(len(X)):
    print(series[i],X[i], y[i])
    count +=1
    if count>7: break
        
# series = np.array([22, 23, 26, 30, 35, 33, 34, 31, 30, 29, 27, 25, 24, 21, 23, 26, 29, 30, 34, 37, 31, 28, 26, 23])        
```
