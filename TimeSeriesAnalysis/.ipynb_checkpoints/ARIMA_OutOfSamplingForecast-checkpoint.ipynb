```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the data
accs = pd.read_csv("data/accounts.csv")
trs = pd.read_csv("data/transactions.csv")
```

```python
# data transformations 
trs['date'] = pd.to_datetime(trs['date'])
t2day = trs['date'].value_counts().sort_values()
t2day = t2day.to_frame()
```

### Using ARIMA for out of sampling forecasts. 

-  Out of sampling forecast is what happens in the real world. (You don't have a test set anymore. )

-  For performing OOS forecast with ARIMA we use the forecast() function.   

-  pay attention to the difference between forecast() and predict() functions

-  to get the forecast invert each forecasted time step, one at a time .
                                                                
-  we add them to the end of a list called history for use when calling inverse_difference() function.

```python
# create a differenced series
def difference(dataset, interval=1):
    diff = list()
    for i in range(interval, len(dataset)):
        value = dataset[i] - dataset[i - interval]
        diff.append(value)
    return np.array(diff)
 
# invert differenced value
def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]
```

```python
X = t2day.date.sort_index().values
days_in_year = 365
differenced = difference(X, days_in_year)
```

```python
# fit model
model = ARIMA(differenced, order=(8,1,2))
model_fit = model.fit()
# multi-step out-of-sample forecast
forecast = model_fit.forecast(steps=30)
# invert the differenced forecast to something usable
history = [x for x in X]
day = 1
predictions = []
for yhat in forecast:
    inverted = inverse_difference(history, yhat, days_in_year)
    print('Day %d: %f' % (day, inverted))
    predictions.append(inverted)
    history.append(inverted)
    day += 1
```

```python
plt.figure(figsize = (20, 6))
plt.xlabel("Time")
plt.ylabel("Amount")

plt.plot(predictions)
plt.show()
```
