```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VARMAX
from sklearn.metrics import mean_squared_error
# VARMA example
from statsmodels.tsa.statespace.varmax import VARMAX
from random import random
```

```python
data2 = pd.read_csv("data/DK-DK2.csv")
data2.head(2)
```

```python
new_cols = []
for i in data2.columns: 
    if i.startswith("latest_forecasted_"):
        new_cols.append(i)

new_cols.append("carbon_intensity_avg")

new_data = data2[new_cols]

for i in new_data.columns: 
    new_data[i] = new_data[i].bfill().ffill()

new_data.drop("latest_forecasted_power_net_import_SE_avg", axis = 1, inplace = True)
new_data.isna().sum()
```

```python
data = new_data[["latest_forecasted_consumption_avg", "carbon_intensity_avg"]]
```

```python
# Plot the time series
data.plot(figsize=(10, 6))
plt.title('Synthetic Time Series Data')
plt.show()
```

```python
data_r = data[41500:].reset_index(drop=True)
data_r.shape
```

```python
data_r
```

```python
# Differencing the data to make it stationary
data_diff = data_r.diff().dropna()

# Plot the differenced series
data_diff.plot(figsize=(10, 6))
plt.title('Differenced Time Series Data')
plt.show()
```

```python
# Split the data into training and test sets
train_size = int(len(data_diff) * 0.95)
train, test = data_diff.iloc[:train_size], data_diff.iloc[train_size:]
print(len(train), len(test))
```

```python
model = VARMAX(train, order=(24, 24))  # You can experiment with different orders
fitted_model = model.fit(disp=False)
```

```python
# Number of steps to forecast
n_forecast = len(test)

# Forecasting multiple steps ahead
forecast = fitted_model.get_forecast(steps=n_forecast)
forecast_df = forecast.predicted_mean

# Revert the differencing to get the forecast in the original scale
forecast_df = train.iloc[-1] + forecast_df.cumsum()

# Combine actual and forecasted data
test_forecast = pd.concat([data_r.iloc[train_size-1:], forecast_df], axis=1)
test_forecast.columns = ['Actual latest_forecasted_consumption_avg', 'Actual carbon_intensity_avg', 'Forecast latest_forecasted_consumption_avg', 'Forecast carbon_intensity_avg']

# Plot the forecast
test_forecast.plot(figsize=(10, 6))
plt.axvline(data_r.index[train_size-1], color='r', linestyle='--')  # Mark the point where forecast starts
plt.title('Multi-step Forecasting')
plt.show()
```

```python
# Calculate the MSE for each time series
mse_x1 = mean_squared_error(test['latest_forecasted_consumption_avg'], forecast_df['latest_forecasted_consumption_avg'])
mse_x2 = mean_squared_error(test['carbon_intensity_avg'], forecast_df['carbon_intensity_avg'])

print(f'Multi-step MSE for X1: {mse_x1}')
print(f'Multi-step MSE for X2: {mse_x2}')
```

```python
mae = np.mean(np.abs((test['latest_forecasted_consumption_avg'] - forecast_df['latest_forecasted_consumption_avg'])))
mae
```
