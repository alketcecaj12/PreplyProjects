```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
import scipy.interpolate as si
```

```python
trs = pd.read_csv("data/transactions.csv")
```

```python
trs['date'] = pd.to_datetime(trs['date'])
```

```python
trs
```

```python
trs_per_acc = trs.groupby('account_id')
```

```python
trs_per_acc_1248 = trs_per_acc.get_group(1248)
```

```python
trs_per_acc_1248_neg = trs_per_acc_1248[trs_per_acc_1248['amount'] < 0]
```

```python
trs_per_acc_1248_neg.tail(20)
```

```python
dates = []
amounts = []
for i, row in trs_per_acc_1248_neg.iterrows(): 
    date_i = row['date']
    a_i = float(row['amount'])
    
    dates.append(date_i)
    amounts.append(a_i)
```

```python
df_inter = pd.DataFrame(amounts, index=dates, columns=['conc'])
```

```python
df_inter.tail(50)
```

```python
f = si.interp1d(pd.to_numeric(df_inter.index), df_inter.conc)
f(pd.to_numeric(pd.to_datetime(['2018-08-14'])))
```

```python
f(pd.to_numeric(pd.to_datetime(['2018-01-07'])))
```

```python
f(pd.to_numeric(pd.to_datetime(['2021-06-29'])))
```
