```python
# DS Basics
import numpy as np
import pandas as pd
#import pandas_profiling
import matplotlib.pyplot as plt

# SKLearn Stuff

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Import data
data = pd.read_csv('data/Demographic_Data.csv')
data.head()
```

```python
age_class = []

for i, row in data.iterrows(): 
    age_i = row['age']
    
    if age_i <= 33: 
        age_class.append(33)
    if age_i > 33 and age_i <= 56: 
        age_class.append(56)
        
        
    if age_i > 56: 
        age_class.append(85)
        
```

```python
data['age_class'] = age_class
data
```

```python
# Select the features
X = data.iloc[:,0:4]
print('Summary of feature sample')
X.head()
```

```python
# dependent variable
y = data['region']
```

```python
# Train/Test splitting data into 70% for training and 30% for testing 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state = 123)
```

```python
#Modeling (Classification)
algo = DecisionTreeClassifier()
model = algo.fit(X_train,y_train)
```

```python
#Predictions
preds = model.predict(X_test)
print(classification_report(y_test, preds))
```

```python
# to visualize the decision tree built using DecisionTreeClassifier()
from sklearn.tree import plot_tree
fig = plt.figure(figsize=(20,10))
tree = plot_tree(model, feature_names=X.columns,class_names=['0', '1', '2', '3'], filled=True)
```

```python
print(fig)
```

```python
print(tree)
```

```python
# Cross Validation and Choosing the Right Model
from sklearn.model_selection import cross_val_score
model = DecisionTreeClassifier()

# cv = number of folds being held out
print(cross_val_score(model, X, y, cv=3))
```

```python
# Modeling Format
algos_Class = []
algos_Class.append(('Random Forest Classifier', RandomForestClassifier()))
algos_Class.append(('Decision Tree Classifier', DecisionTreeClassifier()))
```

```python
#classification
results = []
names = []
for name, model in algos_Class:
    result = cross_val_score(model, X,y, cv=3, scoring='accuracy')
    names.append(name)
    results.append(result)
```
