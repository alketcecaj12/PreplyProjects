```python
import pandas as pd 
import matplotlib.pyplot as plt
import time
data = pd.read_csv("data/data-final_nosso_big5.csv", sep= '\t')
```

```python
for i in data:
    if isinstance(data[i][2], str): continue
    data[i].fillna(data[i].median(),inplace=True)
```

```python
data_clean = data.dropna()
data_clean.isna().sum()
```

```python
X = data_clean.iloc[: , :100]
countries = data_clean["country"].tolist()
countries
```

```python
from sklearn.preprocessing import Normalizer

transformer = Normalizer().fit(X)  # fit does nothing.

X = transformer.transform(X)
```

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
kmeans = MiniBatchKMeans(n_clusters=2,random_state=0,batch_size=512).fit(X)
```

```python
kmeans.inertia_
```

```python
start =  time.time()
print('Start computation time: ', start)
ks = range(1, 40)
in_s = []
for i in ks: 
    model = MiniBatchKMeans(n_clusters=i, random_state=0,batch_size=64)
    model.fit(X)
    
    in_i = model.inertia_
    in_s.append(in_i)
    print(i, in_i)
print('End computation time: ', time.time() - start)
```

```python
plt.plot(ks, in_s, '-o')
plt.xlabel("Number of clusters, k")
plt.ylabel("Error inertia_")
plt.show()
```
