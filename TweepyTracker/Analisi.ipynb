```python
import string
import re
from wordcloud import WordCloud
import matplotlib.pyplot as plt
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct.lower()

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens
```

```python
lista = []
with open('2022-01-12 13_57_46_tweets.txt', 'r', encoding='cp850' ) as file: 
    
    line = file.read()
    lista.append(line)
```

```python
lista 
```

```python
text = "\n-----------------------\n".join(lista)
```

```python
tweets = text.split("\n-----------------------\n")
```

```python
tweets_number = {}
count = 1
for i in tweets: 
    tweets_number[count] = i 
    count+=1
```

```python
tweets_number
```

```python
for k, v in tweets_number.items(): 
    print(k, v)
    text_i = v
    print(type(text_i))
    text_i = remove_punct(text_i)
    print(text_i)
    l = tokenize(text_i)
    print(l)
    print('--------------------------------------------------')
    
```

```python
import nltk

stopword = nltk.corpus.stopwords.words('italian')
```

```python
def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text
```

```python
token_list = []
for k, v in tweets_number.items(): 
    print(k, v)
    text_i = v
    print(type(text_i))
    text_i = remove_punct(text_i)
    print(text_i)
    l = tokenize(text_i)
    print(l)
    l2 = remove_stopwords(l)
    token_list.append(l2)
    print(l2)
    print('--------------------------------------------------')
```

```python
lista_token_m2 = []
for lista in token_list:
    li_i = []
    for li in lista: 
        if len(li) > 2: 
            li_i.append(li)
    lista_token_m2.append(li_i)        
```

```python
lista_token_m2
```

```python
flatten_list = sum(lista_token_m2, [])
```

```python
flatten_list
```

```python
text_wordcloud = " ".join(flatten_list)
```

```python
text_wordcloud
```

```python
wordcloud_grahics = WordCloud(collocations = False).generate(text_wordcloud)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(wordcloud_grahics, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
