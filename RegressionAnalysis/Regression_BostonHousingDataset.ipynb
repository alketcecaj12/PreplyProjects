### The Boston Dataset

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/train.csv")
```

```python
data
```

```python
# check the data types in the data frame
data.info()
```

```python
# check for missing values
data.isna().any()
```

```python
for i in data.columns: 
    print(i, data[i].isna().sum())
```

```python
### Consider only the columns with less than 30 % of nans
cols = []
for i in data.columns: 
    if data[i].isna().sum() < len(data)*0.3: 
            cols.append(i)
   
```

```python
pre_clean_data = data[cols]
pre_clean_data
```

```python
### Replace missing value swith the median

for i in pre_clean_data.columns: 
    
    if isinstance(pre_clean_data[i][2], str): 
        continue
    pre_clean_data[i] = pre_clean_data[i].fillna(pre_clean_data[i].median())
            
```

```python
# check again for missing data
for i in pre_clean_data.columns: 
    print(i, pre_clean_data[i].isna().sum())
```

```python
pre_clean_data.info()
```

```python
columns = ['Id', 'MSZoning', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 
          'Condition2', 'BldgType', 'HouseStyle', 'RoofMatl','RoofStyle', 'Exterior1st', 'Exterior2nd', 'ExterQual',
           'ExterCond', 'Foundation', 'Heating', 'HeatingQC', 'CentralAir', 'KitchenQual', 'Functional', 'PavedDrive', 
          'SaleType', 'SaleCondition', 'Street', 'MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 
          'Electrical', 'GarageType', 'GarageQual', 'GarageCond', 'GarageFinish', 'BsmtFinType2']


clean_data = pre_clean_data.drop(columns, axis = 1)
```

```python
clean_data
```

```python
clean_data.info()
```

```python
X = clean_data.iloc[:, :36].values
y = clean_data.iloc[:, -1].values
```

### Apply Regression using Scikitlearn

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
yhat = linear_model.predict(X_test)
```

### Compute the forecasting measures

```python
errore_prev = abs(yhat - y_test)
```

```python
import numpy as np
mae = np.mean(errore_prev)
```

```python
mae
```

```python
mape = np.mean(errore_prev / y_test) * 100
```

```python
mape
```

#### Cross Validation - 10 Fold cross-validation using scikit-leanr and the r-square of the model

```python
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model, X, y, cv = 10)
```

#### Cross Validation - 10 Fold cross-validation using a simple for loop

```python
bias = []
for i in range(1,11):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = i)
    linear_model = LinearRegression()
    linear_model.fit(X_train, y_train)
    yhat = linear_model.predict(X_test)
    errore_prev = abs(yhat - y_test)
    mae = np.mean(errore_prev)
    mape = np.mean(errore_prev / y_test) * 100
    print('MAE = ', mae, '     MAPE', mape)
    bias.append(mape)
```

```python
# bias
np.mean(bias)
```

```python
# variance 
np.std(bias)
```
